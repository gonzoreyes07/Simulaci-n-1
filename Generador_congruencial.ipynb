```python
from tabulate import tabulate
import matplotlib.pyplot as plt

def function(a, c, m, x0):
    x = (a * x0 + c) % m
    return x

# Parámetros iniciales
x0 = 5
a = 81
c = 89
m = 10**2

# Listas para almacenar resultados
lista = []          # Para los números normalizados
muestra = []        # Para la tabla
seen = set()        # Se añade un conjunto seen para detectar los números normalizados
p_iteracion_repetida = None  # Para guardar la iteración de la primera repetición
p_num_repetido = None      # Para guardar el número repetido
p_aparicion = None        # Para guardar la primera aparición

# Generación de la secuencia
for i in range(1000):
    nuevo = function(a, c, m, x0)
    num_normalizado = nuevo / m

    # Verificar repeticiones y registrar la primera ocurrencia
    if num_normalizado in seen and p_iteracion_repetida is None:
        p_iteracion_repetida = i
        p_num_repetido = num_normalizado
        p_aparicion = lista.index(num_normalizado)

    # Agregar a las listas
    lista.append(num_normalizado)
    seen.add(num_normalizado)
    muestra.append([i, x0, num_normalizado])
    x0 = nuevo

# Reporte del ciclo (si lo hubo)
if p_iteracion_repetida is not None:
    longitud_ciclo = p_iteracion_repetida - p_aparicion
    print(f"Ciclo detectado en la iteración {p_iteracion_repetida}")
    print(f"El número {p_num_repetido} se repitió (primera aparición en iteración {p_aparicion})")
    print(f"Longitud del ciclo: {longitud_ciclo}")
    print(f"Total de números únicos generados: {len(seen)}")
else:
    print("\nNo se detectó ningún ciclo en las 100 iteraciones.")

# Gráficas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(lista, 'b.-')
plt.title('Secuencia Generada')
plt.xlabel('Iteración')
plt.ylabel('Valor')

plt.subplot(1, 2, 2)
plt.hist(lista, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# Tabla
encabezado = ["Iteracion", "Xn", "Xn+1"]
print(tabulate(muestra, headers=encabezado))
```
