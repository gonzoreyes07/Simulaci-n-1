```python
from tabulate import tabulate

x_0 = 1856 # Semilla inicial

table_headers = ["i", "x_n", "x_n^2", "x_n+1"]

table_rows = []

iteration = 1

while True:
  result = x_0 ** 2 # Se eleva al cuadrado nuestro x_0
  result_str = str(result).zfill(8) # Se agrega la función string().zfill() para agregar ceros a la izquierda hasta que el número sea del tamaño que querramos (en este caso de 8 dígitos)
  x_1 = int(result_str[2:6]) # Seleccionamos los dígitos de enmedio
  table_rows.append([iteration, x_0, result, x_1])

  x_0 = x_1 # Actualizmaos los valores para hacer un ciclo
  iteration += 1

  if iteration > 50:
    break

print(tabulate(table_rows, headers=table_headers, tablefmt="grid"))
```
