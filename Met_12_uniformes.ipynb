```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Definimos nuestra función para generar números aleatorios con distribución normal usando el método de las 12 uniformes
def metodo(n, mean=0, variance=1):

    # Generamos 12*n números uniformes
    uniformes = np.random.uniform(0, 1, (n, 12))

    # Sumamos las 12 uniformes y restamos 6 para centrar en 0
    # La varianza de la suma es 12*(1/12) = 1
    muestras_norm = np.sum(uniformes, axis=1) - 6

    # Ajustamos a la media y varianza deseadas
    # X = sigma * Z + mu, donde Z es la normal estandar
    sigma = np.sqrt(variance)
    muestras_norm = sigma * muestras_norm + mean

    return muestras_norm
```

```python
# Definimos nuestras variables
n_muestras = 10000    # Número de muestras
med_teo = 5  # Media teórica
var_teo = 4  # Varianza teórica
desv_teo = np.sqrt(var_teo)  # Desviación estándar teórica
```

```python
# Generamos las muestras
muestras = metodo(n_muestras, med_teo, var_teo)
```

```python
# Cálculo estadísticos
med = np.mean(muestras)
desv = np.std(muestras)
var = np.var(muestras)
```

```python
# Histograma
plt.figure(figsize=(10, 6))
plt.hist(muestras, bins=50, density=True, alpha=0.7, color='skyblue', label='Histograma')
```

```python
# Generamos la curva teórica de la distribución normal
x = np.linspace(med_teo - 4*desv_teo, med_teo + 4*desv_teo, 100)
normal = (1/(desv_teo * np.sqrt(2*np.pi))) * np.exp(-0.5*((x - med_teo)/desv_teo)**2)
plt.plot(x, normal, 'r-', linewidth=2, label='Distribución Normal Teórica')
plt.title('Método de las 12 uniformes')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.grid(True, alpha=0.3)
plt.legend()
```

```python
print(f"Resultados de la simulación ({n_muestras} muestras):")
print(f"Media teórica: {med_teo:.4f}")
print(f"Media empírica: {med:.4f}")
print(f"Error en la media: {abs(med_teo - med):.4f}")
print(f"\nVarianza teórica: {var_teo:.4f}")
print(f"Varianza empírica: {var:.4f}")
print(f"Error en la varianza: {abs(var_teo - var):.4f}")
print(f"\nDesviación estándar teórica: {desv_teo:.4f}")
print(f"Desviación estándar empírica: {desv:.4f}")
print(f"Error en la desviación estándar: {abs(desv_teo - desv):.4f}")
```
