# Método Monte Carlo Crudo (Media Muestral)

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
```

```python
def f(x): # Programamos la función objetivo y le asignamos el nombre de f
  return np.sqrt(np.arctan(x))
```

```python
def MediasMuestrales(a,b,n,f):
  u = np.random.uniform(0,1,n) # Generamos n números aleatorios con distribución uniforme
  G = [] # Creamos una lista para almacenar los números generados
  for i in range (n):
    x = a + u[i] * (b-a) # Calculamos los valores de la función para obetner su promedio
    G.append(f(x))
  return np.mean(G) # Retorna el promedio de los valores obtenidos
```

```python
def MC_MM(a,b,N,n,f): # Definimos otra función para realizar la simulación del método de MC
  I = []
  for i in range(N):
    I.append(MediasMuestrales(a,b,n,f)) # Obtenemos las medias y guardamos en la lista creada anteriormente.
  return I
```

```python
# Definimos nuestras variables

n = 100 # Números aleatorios
N = 10000 # Número de veces que se repetirá el experimento
a = 0
b = 1
```

```python
I = MC_MM(a,b,N,n,f) # Realizamos la simulación
```

```python
mu = np.mean(I) # Calculamos la aproximación a través de la media de los experimentos realizados
desv = np.std(I) # Calculamos la desviación estándar de los experimentos realizados
```

```python
plt.plot(I)
plt.show()
```

```python
print("Intervalo generado: (",mu-desv,";",mu+desv,")")
print("Valor obtenido generado:", mu)
```
