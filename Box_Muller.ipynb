```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def box_muller(med, desv, n):
  # Generamos nuestros números aleatorios uniformes
  U1 = np.array([random.random() for _ in range(n)])
  U2 = np.array([random.random() for _ in range(n)])

  # Realizamos la transformación de Box-Muller
  R = np.sqrt(-2 * np.log(U1))
  theta = 2 * np.pi * U2
  # X = med + desv * R * np.cos(theta)
  # Y = med + desv * R * np.sin(theta)
  X = R * np.cos(theta)
  Y = R * np.sin(theta)
  return X, Y
```

```python
datos = box_muller(0, 1, 10000)
```

```python
# Calculamos la media y desviación estándar
media = np.mean(datos)
desviacion = np.std(datos)

# Imprimimos los resultados
print("Media:", media)
print("Desviación estándar:", desviacion)
```

```python
# Creamos el histograma
plt.figure(figsize=(10, 6))

# Plot each dataset with a different color
plt.hist(datos[0], bins=50, density=True, alpha=0.6, color="blue", label="Histograma X")
plt.hist(datos[1], bins=50, density=True, alpha=0.6, color="green", label="Histograma Y")

x = np.linspace(min(datos[0]), max(datos[0]), 100) # Use datos[0] for x-axis range
curva = 1/(desviacion * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desviacion**2) )
plt.plot(x, curva, "r--", label="Distribución teórica")

plt.title("Método Box-Muller")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.show()
```
